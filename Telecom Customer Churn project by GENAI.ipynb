```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv(r"C:\Users\Prath\Downloads\Customer Churn.csv")

# Display the first few rows of the DataFrame
df.head()
```

```python
# Display basic information about the DataFrame
df.info()
```

```python
# Replace blank spaces with 0 in TotalCharges column
df['TotalCharges'] = df['TotalCharges'].replace(' ', '0')

# Convert TotalCharges column to numeric datatype
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])
```

```python
# checking it's done or not
df.info()
```

```python
#  Check for null values in the dataset
print("Null Values in Dataset:")
print(df.isnull().sum())
```

```python
#  Describe numerical columns
print("\n\nDescription of Numerical Columns:")
print(df.describe())
```

```python
# checking duplicate values
df.duplicated().sum()
```

```python
# 1. Define function to convert value
def convt(value):
    if value == 1:
        return "Yes"
    else:
        return "No"
# Apply this function to SeniorCitizen column
df['SeniorCitizen'] = df['SeniorCitizen'].apply(convt)
        
```

```python
df.head(40)
```

#converted the value of seniorcitizen column with 1=="yes" else no

```python
#count plot for churn customer and uncharn customer
plt.figure(figsize=(8, 6))
ax = sns.countplot(data=df, x='Churn')
plt.title('Count of Customers by Churn Status')
plt.xlabel('Churn Status')
plt.ylabel('Count of Customers')

# Add count values on top of each bar
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

```python
# pie chart for better understanding
churn_counts = df['Churn'].value_counts()
labels = churn_counts.index
sizes = churn_counts.values

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Customers by Churn Status')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

plt.show()
```

```python

# List of columns
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Define subplot grid
n_cols = 3  # number of plots in one row
n_rows = (len(cols) + n_cols - 1) // n_cols  # rows needed

plt.figure(figsize=(15,15))  # Adjust figure size

for i, col in enumerate(cols, 1):
    plt.subplot(n_rows, n_cols, i)  # create subplot
    sns.countplot(data=df, x=col, hue ="Churn")
    plt.title(col, fontsize=12)
    plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```
